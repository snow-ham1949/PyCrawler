### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

NEWS = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[0].text, d.find_all('a')[-1].text)
    news = (d.find(class_="date")).text, d.find_all('em')[0].text, d.find_all('a')[-1].text
    NEWS.append(news)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(NEWS, columns=["date","category","title"])

print(data)
```

### ③ 哪一個類別的文章最多

```python
data["category"].value_counts(ascending=False)
```
